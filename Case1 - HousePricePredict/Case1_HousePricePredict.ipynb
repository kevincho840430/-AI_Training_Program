# 建模 - 線性迴歸模型

這裡要比較對數轉換前後的差異

`LinearRegression`：最基礎的線性迴歸模型，目的在於預測房價的多寡。

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

 `lm.fit`：`fit`訓練模型函數

```python
#先將剛新增的非數值型資料剔除
df = df.drop(columns = ['AGE_new','RM_new'])
```

```python
lm.fit(X = df.iloc[:,:-1], y = df.iloc[:,-1])
```

`lm.score`：計算方式是使用`df.iloc[:,:-1]`預測，並與目標房價`df.iloc[:,-1]`做比較，得到分數。

$R-square$模型評價分數，分數越高越好，值域為$(0,1]\$，此分數表示模型中有多少`variance`可以被特徵所解釋。

這裡分數約為$0.74$，表示特徵可以在此模型解釋百分之$74\%$的`variance`。

```python
lm.score(df.iloc[:,:-1], df.iloc[:,-1])
```
