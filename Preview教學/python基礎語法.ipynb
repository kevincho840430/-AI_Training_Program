## 條件判斷式 (if else)
- 當我們需要進行多個條件的判斷的時候就可以用if else語法判斷
- 可以很直觀的想成如果達成了某個條件就會進行以下動作，不然則會執行另外一個動作。

```python
# 條件判斷可以用 and 或 or 來判斷多個條件
score = 90
if(score >= 60 and score < 70): # 條件判斷式以 if 做開頭
  print("you are C!")
elif(score >= 70 and score < 80): # if 後面的條件都用 elif 
  print("you are B!")
elif(score >= 80 and score < 90):
  print("you are A!")
elif(score >= 90 and score <= 100):
  print("you are S!")
else: # 最後用 else 結尾，前面的條件都沒成立的話，就會進到這邊
  print("you are fail!")
```

## 函數
- 當一段程式我們需要重複使用或呼叫的時候，就可以用def function的寫法重複呼叫
- 在寫函式的時候我們需要將原本print的地方改為return，原因是因為叫function的時候我們需要將結果回傳出來才能使用。

```python
#延續上面的分數判斷將其改成def function寫法
def grade(score):
  if(score >= 60 and score < 70): # 條件判斷式以 if 做開頭
    return ("you are C!")
  elif(score >= 70 and score < 80): # if 後面的條件都用 elif 
    return ("you are B!")
  elif(score >= 80 and score < 90):
    return ("you are A!")
  elif(score >= 90 and score <= 100):
    return ("you are S!")
  else: # 最後用 else 結尾，前面的條件都沒成立的話，就會進到這邊
    return ("you are fail!")

print(grade(59))
print(grade(70))
```

## Range
Range這個型別可用來創建並儲存特定範圍內的整數，故得名Range。
必須特別注意的是，一旦Range被創建了，裡面的內容是不可修改的。
在Python中，我們有幾個方法可以創造Range。
- Range(stop)
stop：停止點
- Range(start, stop)
start：起始點
stop：停止點
- Range(start, stop, step)
start：起始點
stop：停止點
step：間隔

- 通常搭配for迴圈或List來使用

```python
r1 = range(10)
r2 = range(5, 50, 5)

print(type(r1))

print(list(r1))
print(list(r2))
```

- 若沒有給起始值，將預設為0
- 若沒有給間隔，將預設為1
- 遇到停止點後，創造的過程就會終止，因此Range中的數字將不會包含停止點

## while迴圈
- 是用條件判斷去決定跑幾圈

```python
#重新再跑一次
def grade(score):
  if(score >= 60 and score < 70): # 條件判斷式以 if 做開頭
    return ("you are C!")
  elif(score >= 70 and score < 80): # if 後面的條件都用 elif 
    return ("you are B!")
  elif(score >= 80 and score < 90):
    return ("you are A!")
  elif(score >= 90 and score <= 100):
    return ("you are S!")
  else: # 最後用 else 結尾，前面的條件都沒成立的話，就會進到這邊
    return ("you are fail!")
```

- 提示二: 設一變數 sum = 0, 再來 sum = sum + 某數，用這種寫法就可以達到累加的效果

```python
temp_list = [23,65,78,23,14]

# 請將 function 完成
def even_number_sum(number_list):

#

print(even_number_sum(temp_list))
```

## 解答

```python
#定義一 list
temp_list = [23,65,78,23,14]
#定義even_number_sum函數
def even_number_sum(number_list):
  #指定一常數sum為0，將其用來儲存總和
  sum = 0
  #使用for迴圈跑除以2餘數為0的數字，並將其加到sum這個值裡面
  for number in number_list:
    if(number % 2 == 0):
      sum = sum + number
  #回傳加總後的數字
  return sum
print(even_number_sum(temp_list))
```

## try except
- 當程式執行時發現有錯誤或者是需要進行符合條件時才執行的狀況可使用try except
- 如果發生例外，程式流程會跳離例外發生點，比對except宣告的型態

```python
#right
image = [(1,2,3),(4,5,6)]
for i in image:
    print(i)
```

```python
#error
image = None
for i in image:
    print(i)
```

```python
#解答
try: 
    image = None
    for i in image:
        print(i)
except:
    print('Nothing')
```
